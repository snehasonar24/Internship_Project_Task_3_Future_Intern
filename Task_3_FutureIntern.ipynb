<span style="text-align: center;color:Blue;text-align: center;font-weight:bold ; font-size:20px">Create a histogram or bar chart to visualize the distribution of data in a dataset</span>



```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r"D:\Future Intern\Iris.csv")
```

```python
numeric_columns = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
```

<span style="text-align: center;color:Yellow;font-size: 20px;text-align: center;font-weight:bold ; font-size:40px">Plot histograms for each numeric column</span>

```python
plt.figure(figsize=(12, 12))
for i, column in enumerate(numeric_columns, 1):
    plt.subplot(2, 2, i)  # Create a 2x2 grid for the plots
    plt.hist(data[column], bins=10, color='blue', edgecolor='black')  # Create the histogram for each column
    plt.title(f'Distribution of {column}')  # Add a title for each subplot
    plt.xlabel(column)  # Label the x-axis
    plt.ylabel('Frequency')  # Label the y-axis


plt.tight_layout()
plt.show()
```

<span style="text-align: center;color:Yellow;font-size: 20px;text-align: center;font-weight:bold ; font-size:40px">Plot Bar Charts</span>

```python
#Group the data by 'Species' and calculate the mean for each group
grouped_data = data.groupby('Species').mean()

#  Plot bar charts for each numeric column
plt.figure(figsize=(12, 10)) 

for i, column in enumerate(grouped_data.columns[:4], 1):  # Restrict to first 4 columns
    plt.subplot(2, 2, i)  
    plt.bar(grouped_data.index, grouped_data[column], color='green', edgecolor='black')  
    plt.title(f'Average {column} by Species') 
    plt.xlabel('Species')  
    plt.ylabel(f'Average {column} (cm)')  

plt.tight_layout()
plt.show()
```

```python

species_count = data['Species'].value_counts()

# Plot a bar chart for the distribution of species
plt.figure(figsize=(8, 6))
species_count.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title('Distribution of Iris Species in the Dataset')
plt.xlabel('Species')
plt.ylabel('Count')
plt.xticks(rotation=0)  


plt.show()
```
